```python
# 모듈 import
import requests
import pprint
import xml.etree.ElementTree as ET
#인증키 입력
encoding = 'Vmb9cFZspt9fIixQRRVb7D9iKDa%2BLFWo6Z3ED8LUGYEynog9GobKv8DEsgpl9ieXYFvqGXbnbID1bPWz2%2BpnDQ%3D%3D'
decoding = 'Vmb9cFZspt9fIixQRRVb7D9iKDa+LFWo6Z3ED8LUGYEynog9GobKv8DEsgpl9ieXYFvqGXbnbID1bPWz2+pnDQ=='

#url 입력
url = 'http://apis.data.go.kr/B554287/DisabledPersonConvenientFacility/getDisConvFaclList'
params ={'serviceKey' : encoding , 
'pageNo' : '1', 
'numOfRows' : '1000'}

response = requests.get(url, params=params)

print(response.content)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

# XML 데이터 파싱
root = ET.fromstring(response.content)

# servList 요소들 찾기
serv_list_nodes = root.findall('.//servList')

# 각 서비스 정보를 딕셔너리 리스트로 저장
data = []
for serv_list_node in serv_list_nodes:
    estbDate = serv_list_node.find('estbDate')
    faclInfId = serv_list_node.find('faclInfId')
    faclLat = serv_list_node.find('faclLat')
    faclLng = serv_list_node.find('faclLng')
    faclNm = serv_list_node.find('faclNm')
    faclRprnNm = serv_list_node.find('faclRprnNm')
    faclTyCd = serv_list_node.find('faclTyCd')
    lcMnad = serv_list_node.find('lcMnad')
    salStaDivCd = serv_list_node.find('salStaDivCd')
    salStaNm = serv_list_node.find('salStaNm')
    wfcltDivCd = serv_list_node.find('wfcltDivCd')
    wfcltId = serv_list_node.find('wfcltId')
    
    data.append({
        "Establishment Date": estbDate.text if estbDate is not None else None,
        "Facility Info ID": faclInfId.text if faclInfId is not None else None,
        "Facility Latitude": faclLat.text if faclLat is not None else None,
        "Facility Longitude": faclLng.text if faclLng is not None else None,
        "Facility Name": faclNm.text if faclNm is not None else None,
        "Facility Representative Name": faclRprnNm.text if faclRprnNm is not None else None,
        "Facility Type Code": faclTyCd.text if faclTyCd is not None else None,
        "Facility Basic Address": lcMnad.text if lcMnad is not None else None,
        "Sales Status Division Code": salStaDivCd.text if salStaDivCd is not None else None,
        "Sales Status Name": salStaNm.text if salStaNm is not None else None,
        "Welfare Facility Management Division Code": wfcltDivCd.text if wfcltDivCd is not None else None,
        "Welfare Facility ID": wfcltId.text if wfcltId is not None else None,
    })

# 데이터프레임으로 변환
df = pd.DataFrame(data)

# 데이터프레임 출력
df
```

```python
# 각 행의 컬럼, 이름, 값을 가지는 리스트 만들기
row_list = [] # 행값
name_list = [] # 열이름값
value_list = [] #데이터값

# xml 안의 데이터 수집
for i in range(0, len(rows)):
    columns = rows[i].find_all()
    #첫째 행 데이터 수집
    for j in range(0,len(columns)):
        if i ==0:
            # 컬럼 이름 값 저장
            name_list.append(columns[j].name)
        # 컬럼의 각 데이터 값 저장
        value_list.append(columns[j].text)
    # 각 행의 value값 전체 저장
    row_list.append(value_list)
    # 데이터 리스트 값 초기화
    value_list=[]

#xml값 DataFrame으로 만들기
corona_df = pd.DataFrame(row_list, columns=name_list)
###assertion error의 경우###
###corona_df = pd.DataFrame(row_list)
print(corona_df.head(19)) 
```
